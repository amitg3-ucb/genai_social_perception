# **Cleaned Data**

Data above is really tough to work with, going to clean text data to remove symbols, and put everything in lowercase to eventually tokenize words


```python
# Function to clean data with numbers

def  clean_text_with_numbers(df_original: df, text_field: str, new_text_field_name: str):
    df = df_original.copy()
    df[new_text_field_name] = df[text_field].str.lower()
    df[new_text_field_name] = df[new_text_field_name].apply(lambda elem: re.sub(r"(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", "", elem))
    return df

# Function to clean data without numbers
def  clean_text_without_numbers(df_original: df, text_field: str, new_text_field_name: str):
    df = df_original.copy()
    df[new_text_field_name] = df[text_field].str.lower()
    df[new_text_field_name] = df[new_text_field_name].apply(lambda elem: re.sub(r"(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", "", elem))
    # remove numbers
    df[new_text_field_name] = df[new_text_field_name].apply(lambda elem: re.sub(r"\d+", "", elem))
    return df
```

```python
tweet_clean_no_numbers = clean_text_without_numbers(df,'Text','clean_tweet')
tweet_clean_no_numbers
```

```python
tweet_clean_with_numbers = clean_text_with_numbers(df,'Text','clean_tweet_numbers')
tweet_clean_with_numbers
```

```python
# Function to filter stop words
def filter_stop_words(text):
    words = nltk.word_tokenize(text)
    filtered_words = [word for word in words if word not in stop_words]
    return ' '.join(filtered_words)
```

```python
reset_df['filtered_text'] = reset_df['clean_tweet'].apply(filter_stop_words)

# Print the updated DataFrame
reset_df
```

```python
test_text_stopwords = reset_df['filtered_text'][0]
test_text_stopwords
```

```python
sentiment_scores = analyzer.polarity_scores(test_text_stopwords)
print(sentiment_scores)
```

```python
def get_sentiment_score(test_text_stopwords: str):
  return analyzer.polarity_scores(test_text_stopwords)
```

```python
# Apply sentiment score and create new column of top 30 tweet counts
reset_df['sentiment_score'] = reset_df['filtered_text'].apply(get_sentiment_score)

reset_df
```

#Below is data processing to get total count of Tweets using datetime column




```python
string_datetime = df['Datetime'].tolist()

string_datetime[0:5]
```

```python
# This is the way we want to format the date
format_date = '%a %b %d %H:%M:%S %z %Y'

converted_date = []

# Loop through each date and format
for date_string in string_datetime:
    converted_datetime = datetime.strptime(date_string, "%Y-%m-%d %H:%M:%S%z").replace(tzinfo=timezone.utc)
    converted_date.append(converted_datetime)
converted_date[0:5]
```

```python
tweet_counts = {}
#Loop through converted dates and find DOW and HOR
for dates in converted_date:
    day = dates.strftime('%A')
    hour = dates.strftime('%H')

    # Create day of week and hour of day key value pair here
    key = (day, hour)

    # Update tweet counts dictionary
    if key in tweet_counts:
        tweet_counts[key] += 1
    else:
        tweet_counts[key] = 1
```

```python
# Create data to graph
graph_data = []
for key, count in tweet_counts.items():
    day, hour = key
    graph_data.append({'day': day, 'hour': hour, 'count': count})

print(graph_data[0:5])
```

```python
graph_df = pd.DataFrame(graph_data)
y_order = ['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']
graph_df
```
